```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import scipy.optimize as sco
import scipy.constants as c
from scipy.optimize import curve_fit as cf
import pandas as pd
```

```python
def gauss(x,mu,sig,A,b):
    return A*np.exp( - (x-mu)**2 / 2 / sig**2) +b
```

```python
def plotGraph(ordner, anzahl, xoff=0, yoff=0, g=False, visible=True):
    if visible:
        fig,ax=plt.subplots(dpi=144)
        ax.plot([],[]," ", label=ordner)
    if g: rangee= [3,4,5]
    else: rangee = range(anzahl)
        
    data_array=[]
    for n in rangee:
        xShift,yShift=0,0
        data=pd.read_csv("Data/{}/ALL000{}/F000{}CH1.csv".format(ordner,n,n),skiprows=17)
        
        if n==rangee[-1]: xShift,yShift = xoff,yoff
        t=np.array(data.values[:,3]).astype(float) 
        intens=np.array(data.values[:,4]).astype(float) + yShift

        t=t-min(t) # Damit die Skala bei 0 realtiv anfngt -- keine neg Energien
        time= t*1000 + xShift
        data_array.append(time)
        data_array.append(intens)
        
        if visible==False: continue
        ax.plot(time,intens, label="{}".format(n))
        ax.set(yticks=[],ylabel="$Intenstiy\; [a.u.]$", xlabel="$Time\; [ns]$"
               , xticks=np.linspace(0,10*round(max(time)/10),1+int(10*round(max(time)/10)/2.5)))
    
    if visible:
        ax.legend()
        ax.grid(which="Both")
    
    return data_array # time,intens , time,intens , ... -> [i],[i+1] to call
    
#plotGraph("87F2", 3, g=True) # andere 3 
plotGraph("87F2", 3, xoff=.65, yoff=-1)
```

```python
# Fit for the sidebands to get an energy 

def find_cloeset_Pos_to_Value(array,value):
    temp=0
    count=0
    for i in range(len(array)):
        if temp < value:
            temp=array[i]
            count+=1
            continue
        return int(count)  

def gaussFit(time, intens, mu_guess,width_list,ordner):
    mu_list,dmu_list=[],[]
    time_range=np.linspace(min(time),max(time),len(time))
    
    fig,ax=plt.subplots(dpi=144)
    ax.plot(time,intens,alpha=.4)
    ax.plot([],[]," ", label=ordner)
    
    for i in range(len(mu_guess)):   
        print("found one")
        peak_pos=find_cloeset_Pos_to_Value(time,mu_guess[i])
        
        width=width_list[i]
        fit_width=time_range[peak_pos-width[0] : peak_pos+width[1]]
        popt,pcov = cf(gauss, fit_width, intens[peak_pos-width[0] : peak_pos+width[1]], p0=[mu_guess[i], 1, 1, 1])
        ax.plot( fit_width, gauss(fit_width,*popt),label=i+1)
        mu_list.append(popt[0])
        dmu_list.append(abs(popt[1])) # sigma als fehler auf mu
    
    ax.set(yticks=[],ylabel="$Intenstiy\; [a.u.]$", xlabel="$Time\; [ns]$")
    ax.legend()
    return mu_list,dmu_list
    
data= plotGraph("87F2", 3, xoff=.0007, yoff=-1,visible=False)
dataSet=4
mu_guess= [10,13,17,20.5,22.9, 26.9,32,36,39]
width_list=[[50,50] for i in mu_guess]
gaussFit(data[dataSet], data[dataSet+1]-data[0+1], mu_guess ,width_list,"87F2")
```

```python
name="85F3"
data= plotGraph(name, 2,yoff=-.1,visible=True)
dataSet=0
print(data[dataSet][840])
mu_guess= [1.7,3.2,4.3,5.85]
width_list=[[70,70],[70,100],[100,70],[70,35]]
gaussFit(data[dataSet], data[dataSet+1]-data[1+1]*0.15, mu_guess ,width_list,name)
```

```python
name="85F2"
data= plotGraph(name, 2,yoff=-.05,visible=True)
dataSet=0
print(data[dataSet][1470])
mu_guess= [3.6,4.8,5.35,6.3]
width_list=[[70,70],[200,10],[50,160],[0,80]]
gaussFit(data[dataSet], data[dataSet+1]-data[1+1]*0.15, mu_guess ,width_list,name)
```
