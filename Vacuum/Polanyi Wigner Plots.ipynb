```python
%matplotlib inline
import numpy as np
import scipy as sci
import scipy.constants as const
import matplotlib.pyplot as plt
```

```python
kb=const.Boltzmann
R=const.gas_constant
h=const.h
print(R)
def jdes(E,T,n):
    return (kb*T/h*n*np.exp(-E/(R*T)))
```

```python
mylin=np.linspace(0,60000,2000)
trange=[300,500,800,1000]
print(trange[0])
y=[]
for temp in trange:
    y.append(jdes(mylin,temp,5.35*(10**13)))

fig, axes = plt.subplots(figsize=(3,3),dpi=300)
for x in range(len(y)):
    axes.plot(mylin, y[x],label=str(trange[x])+"K")
axes.legend()
axes.set_title("Polanyi Wigner")
axes.set_xlabel(r"$E_{des}$ [J]")
axes.set_ylabel(r"$j_{des}$")
axes.set_ylim()
```
