```python
%matplotlib inline
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
import pandas as pd
```

```python

def plotcorr(name,ax):
    data=np.genfromtxt(name,skip_header=16,delimiter="\t",filling_values=-1,comments="\"")
    time=data[:183,0]
    corr=data[:183,1]

    ax.plot(time,np.log(np.abs(corr)),label=name)
    ax.set_title(name)
```

```python
#Alles in einen plot :D
namelist=[]
for x in range(2,16):
     namelist.append("Data/A/{}.ASC".format(x*10))
fig,ax=plt.subplots(dpi=144)
for name in namelist:
    plotcorr(name,ax)
ax.legend()
```

```python

def plotcorr(name):
    data=np.genfromtxt(name,skip_header=16,delimiter="\t",filling_values=-1,comments="\"")
    time=data[:183,0]
    corr=data[:183,1]
    fig,ax=plt.subplots(dpi=144)
    ax.plot(time,np.log(np.abs(corr)),".",label=name)
    ax.set_title(name)
    fig.savefig()
```

```python
#Alles in einen plot :D
namelist=[]
for x in range(2,16):
     namelist.append("Data/A/{}.ASC".format(x*10))

for name in namelist:
    plotcorr(name,)
ax.legend()
```
