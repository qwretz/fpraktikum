```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import scipy.optimize as sco
import scipy.constants as c
from scipy.optimize import curve_fit as cf
import pandas as pd
```

```python
#Brechungsindex
n=1.37927
#Laserwellenlänge
wavelen=532e-9
```

```python
def plotI(name,labels,ax):
    data=np.genfromtxt(name,skip_header=16,delimiter="\t",filling_values=-1,comments="\"")
    time=data[183:,0]
    I=data[183:,1]
    #sci.optimize.curve_fit(g,time,corr,)
    
    
    ax.plot(time,I,label=r"{} $^\circ$".format(labels))
    ax.set_title(name)
    
#Alles in einen plot :D
namelist=[]
angles=[]
for x in range(2,16):
    namelist.append("Data/A/{}.ASC".format(x*10))
    angles.append(x*10)
    

#fig,ax=plt.subplots(dpi=144)
#plotI(namelist[0],angles[0],ax)
```

```python
gamma= np.pi/180 # Umrechnung ins grad Maß
def formfaktor(q,a):
    x=q*a
    return 9 * ( np.sin(x*gamma) - x* np.cos(x*gamma) )**2 /  x**3
```

```python
angles=[]
meanI=[]

for x in range(2,16):
    name="Data/A/{}.ASC".format(x*10)
    angles.append(x*10)
    data=np.genfromtxt(name,skip_header=16,delimiter="\t",filling_values=-1,comments="\"")
    I=data[183:,1]
    meanI.append(np.mean(I) )
    
momentum= [4*np.pi*n/wavelen * np.sin(x/180*np.pi/2) for x in angles]

popt,pcov=cf(formfaktor, momentum, meanI, p0=(500e-9))
perr = np.sqrt(np.diag(pcov))
print(popt)

q=np.linspace(momentum[0],momentum[-1],1000)

fig,ax=plt.subplots(dpi=144)
ax.plot(momentum, meanI ,label="Data")
ax.plot(q, formfaktor(q,popt) ,label="Fit: a= {:.2} $\pm$ {:.2} nm".format(*popt*1e9,*perr*1e9))
ax.set(xlabel=r"Momentumtransfer $q \: [1/m]$", ylabel="Formfactor $P(q)$", title="Formfactor for Data A")
ax.legend();
```
