```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import scipy.optimize as sco
import scipy.constants as c
from scipy.optimize import curve_fit as cf
import pandas as pd
```

```python
def gauss(x,Amp,mu,sig,b): 
    return Amp* np.exp(-(x-mu)**2/2/sig**2) + b
```

```python
def plotFilter(name, guess, getData=False):
    data=[]
    data.append( pd.read_csv("Data/{}.txt".format(name),delimiter="\t",skiprows=1 ))

    wavelen, intens =[],[]
    for entry in data:
        wavelen= entry.values[:,0]
        intens = entry.values[:,1]
        
    if getData==True:
        return wavelen, intens

    popt,pcov=cf(gauss, wavelen,intens,p0=guess)
    perr=np.sqrt(np.diag(pcov))

    chi2= sum([ (intens[i]-gauss(wavelen[i],*popt))**2 / gauss(wavelen[i],*popt) for i in range(len(intens))])

    xdata=np.linspace(wavelen[0],wavelen[-1],1000)
    fig,ax=plt.subplots(dpi=144)
    ax.plot(wavelen,intens, label="Data: {}".format(name))
    ax.axvline(int(name[:-2]), color="lawngreen",alpha=.5)
    ax.plot(xdata,gauss(xdata,*popt), 
            label="Gaussian: \n A={o[0]:.2f}$\pm${e[0]:.2f} \n $\mu$={o[1]:.2f}$\pm${e[1]:.2f} \n $\sigma$={o[2]:.2f}$\pm${e[2]:.2f} \n b={o[3]:.2f}$\pm${e[3]:.2f} \n $\chi^2$={c:.2f}".format(o=popt,e=perr,c=chi2))
    ax.set(xlabel="$\lambda \ [nm]$", ylabel="$Intensity \ [mV]$")
    ax.legend(frameon=False);
    fig.savefig("Plots/{}-Filter.png".format(name))
    
    return [popt[1],perr[1]]
    
filters = ["768nm","905nm","1060nm"]
lit=np.array([768,905,1060])
p0=[[400,lit[0],5,20],[400,lit[1],5,20],[80,lit[2]+10,5,10]]
mu, dmu=[],[]
for i in range(len(filters)):
    temp = plotFilter(filters[i],p0[i])
    mu.append(temp[0])
    dmu.append(temp[1])
mu=np.array(mu)
dmu=np.array(dmu)
```

```python
gerade= lambda x,a,b: a*x+b
popt,pcov=cf(gerade, mu,lit, sigma=dmu, p0=[1.1,0])
perr=np.sqrt(np.diag(pcov))

xrange=np.linspace(lit[0],lit[-1],1000)
fig,ax=plt.subplots(dpi=144)
ax.errorbar(mu,lit,fmt=" ",yerr=dmu,marker=".")
ax.plot(xrange,gerade(xrange,*popt), label="Linear Fit: a={o[0]:.2f}$\pm${e[0]:.2f}, b={o[1]:.2f}$\pm${e[1]:.2f}".format(o=popt,e=perr))
ax.set(ylabel="Filter Values [nm]",xlabel="Values of Gaussian Fit [nm]")
ax.legend()
fig.savefig("Plots/LambdaCorrection.png");
```

```python
import tex

header=[]
for i in filters:
    header.append("{}: $\lambda$ [nm]".format(i[:-2]))
    header.append("I [mV]")

allData=[]
for i in range(len(filters)):
    temp,tempp = plotFilter(filters[i],p0[i],getData=True)
    allData.append(temp)
    allData.append(tempp)
    
tex.get_table(header,allData)    
```
