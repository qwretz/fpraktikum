```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import scipy.optimize as sco
import scipy.constants as c
from scipy.optimize import curve_fit as cf
import pandas as pd
```

```python
names=["Lamp1","Lamp2","Lamp3"]
titles=["Before first measurement.","Between the measurments.", "After second measurement."]

def lampSpec(name,title,plot=True):
    data=[]
    data.append( pd.read_csv("Data/{}.txt".format(name),delimiter="\t",skiprows=0 ))

    wavelen, intens =[],[]
    for entry in data:
        wavelen= entry.values[:,0]
        intens = entry.values[:,1]
    
    if plot==True:
        fig, ax=plt.subplots(dpi=144)
        ax.plot(wavelen,intens,label="{}".format(title))
        ax.set(xlabel="$\lambda \ [nm]$", ylabel="$Intensity \ [mV]$")
        ax.legend()
    
    return wavelen,intens


fig,ax=plt.subplots(dpi=144)
for i in range(len(names)):
    a,b=lampSpec(names[i],titles[i], plot=False)
    ax.plot(a,b,label="{}".format(title[i]),alpha=.7)
    ax.set(xlabel="$\lambda \ [nm]$", ylabel="$Intensity \ [mV]$")
    
ax.legend()
fig.savefig("Plots/All-Lamp-Spectra.png");
```
