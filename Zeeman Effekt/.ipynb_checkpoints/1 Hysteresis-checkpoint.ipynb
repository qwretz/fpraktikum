```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import scipy.optimize as sco
import scipy.constants as c
from scipy.optimize import curve_fit as cf
import pandas as pd
```

```python
def poly3(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d
def dpoly3(x,dx,a,b,c):
    return (3*a*x**2 + 2*b*x + c ) *dx 
```

```python
data=pd.read_table("Data/Mag.txt")
current= np.array(data.values[:,0]).astype(float) 
B= np.array(data.values[:,1]).astype(float) 
dB= np.array(data.values[:,2]).astype(float) # nur Ablesegenauigkeit!

fig, ax=plt.subplots(dpi=144)
ax.errorbar(current,B, yerr=dB,ecolor="red",ls=" ",lw=1,marker=".",markersize=2.5)
ax.set(xlabel="$Current\ [A]$", ylabel="$Magnetic\ Field\ [mT]$")

names=["$B_{up}$","$B_{down}$"]
xrange=np.linspace(min(current),max(current),1000)
for i in [0,1]:
    cur,b,db = current[0+i*10:11+i*10], B[0+i*10:11+i*10], dB[0+i*10:11+i*10]
    popt,pcov=cf(poly3, cur, b, sigma=db )
    perr=np.sqrt(np.diag(pcov))
    
    chi2 = sum([(b[i] - poly3(cur[i],*popt))**2/poly3(cur[i],*popt) for i in range(len(cur)) ])

    ax.plot(xrange,poly3(xrange,*popt),label="Fit for {}".format(names[i]))
    ax.text(30,300-i*200, "Parameter {}:\n a= {o[0]:.4f}$\pm${e[0]:.4f}\n b= {o[1]:.4f}$\pm${e[1]:.4f}\n c= {o[2]:.4f}$\pm${e[2]:.4f}\n d= {o[3]:.4f}$\pm${e[3]:.4f}\n $\chi^2$={x:.2f}".format(names[i],o=popt,e=perr,x=chi2),fontsize=7)
    
ax.legend(frameon=False)
fig.savefig("Plots/hysteresis.png")

print(poly3(9,*popt), dpoly3(9,1,*popt[:-1]))
print(poly3(18,*popt), dpoly3(18,1.5,*popt[:-1]))
print(poly3(32,*popt), dpoly3(32,1,*popt[:-1]))
```

```python
b,db = poly3(9,*popt)*1e-3, dpoly3(9,1,*popt[:-1])*1e-3
lam1, dlam1 = 1065.59e-9 , 43.37e-9
lam2,dlam2 = 1085.33e-9, 49.86e-9

eme= 2*np.pi*c.c/b *(1/lam1 - 1/lam2) 
deme = 2*np.pi*c.c/b * np.sqrt( (db/b*(1/lam1 - 1/lam2))**2 + (dlam1/lam1**2)**2 + (dlam2/dlam2**2)**2 )

eme*1e-14,deme*1e-14
```
